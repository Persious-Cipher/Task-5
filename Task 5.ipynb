```python
import pandas as pd
df = pd.read_csv("train.csv")  # adjust the path if needed
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.head()
df
```

```python
df.head()           # First 5 rows
df.info()           # Column info and nulls
df.describe()       # Summary statistics
df.columns          # List of column names
df.isnull().sum()   # Check for missing values
df.dtypes           # Check data types
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Sex', data=df)
plt.title("Count of Passengers by Sex")
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Count by Class")
plt.show()
```

```python
sns.countplot(x='Embarked', data=df)
plt.title("Passenger Count by Embarkation Port")
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.xticks([0, 1], ['Not Survived', 'Survived'])  # optional, improves readability
plt.show()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df['Fare'], kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
sns.countplot(x='SibSp', data=df)
plt.title("Number of Siblings/Spouses Aboard")
plt.show()
```

```python
sns.countplot(x='Parch', data=df)
plt.title("Number of Parents/Children Aboard")
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Rate by Sex")
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title("Survival Rate by Embarkation Port")
plt.legend(labels=['Not Survived', 'Survived'])
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age Distribution by Survival")
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Fare Distribution by Survival")
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare (Colored by Survival)")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.show()
```

```python
# Only use numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation and plot heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
